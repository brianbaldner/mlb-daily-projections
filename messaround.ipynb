```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
pd.options.plotting.backend = "plotly"
```

```python
data = pd.read_csv("result.csv", encoding="utf-16").drop(["Unnamed: 0"], axis=1)
data = data[data["PA"] > 100]
data
```

```python
pot = data.plot(x="H / 5ab", y="B / 5ab", color='PA', kind="scatter", hover_data={"Name": True, "Net Matchup vs Avg (hits)" : True, "Net Matchup vs Avg (bases)" : True, "Opposing Pitcher" : True,})
pot.show(renderer="vscode")
```

```python
def avg_best_players(x):
    """Calculates the average of the top N values in a Series."""
    return x.sort_values(ascending=False).head(9).mean()
```

```python
pitchers = data[["Opposing Pitcher", "H / 5ab", "B / 5ab"]].groupby("Opposing Pitcher").aggregate(avg_best_players).sort_values("B / 5ab", axis=0)
pitchers = pitchers.reset_index()
pitchers
```

```python
pot = pitchers.plot(x="H / 5ab", y="B / 5ab", kind="area", hover_data={"Opposing Pitcher": True})
pot.show(renderer="browser")
```
